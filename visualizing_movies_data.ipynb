```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset (assuming the CSV file is in the same directory)
df = pd.read_csv('movies_dataset.csv')
```

```python
# Display the first few rows of the dataset
df.head()
```

```python
#Conclusion: The dataset has been successfully loaded and the first few rows are displayed, providing a quick overview of the data structure, including the columns and sample data.
```

```python
# Get a summary of the DataFrame
df.info()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Basic statistical description of the dataset
df.describe()
```

```python
#Conclusion:df.info(): This provides a summary of the dataset, including the number of entries, the data type of each column, and the memory usage. It shows if there are any missing values and helps in understanding the overall structure of the dataset.
#df.isnull().sum(): This command checks for missing values in each column. If the sum is 0 for all columns, it indicates that there are no missing values in the dataset.
#df.describe(): This gives a statistical summary of the numerical columns in the dataset, including the count, mean, standard deviation, minimum, and maximum values. It helps in understanding the distribution and range of the data.
```

```python
#a. Visualize the Count of Movies by Genre
plt.figure(figsize=(10, 6))
sns.countplot(y=df['genres'], order=df['genres'].value_counts().index)
plt.title('Count of Movies by Genre')
plt.xlabel('Count')
plt.ylabel('Genres')
plt.show()
```

```python
#Conclusion: This visualization shows the distribution of movies across different genres. The genres with the highest and lowest number of movies can be easily identified, helping to understand which genres are more prevalent in the dataset.
```

```python
#b. Vote Count by Genre
grouped_data = df.groupby('genres')['vote_count'].sum().sort_values()

plt.figure(figsize=(10, 6))
grouped_data.plot(kind='barh', color='skyblue')
plt.title('Total Vote Count by Genre')
plt.xlabel('Total Vote Count')
plt.ylabel('Genres')
plt.show()
```

```python
#Conclusion: This bar chart shows the total number of votes each genre has received. It helps identify which genres are more popular based on vote counts. Genres with the highest vote counts indicate a higher level of audience engagement or popularity.
```

```python
#c. Distribution of Vote Average
plt.figure(figsize=(8, 6))
plt.hist(df['vote_average'], bins=10, color='lightgreen', edgecolor='black')
plt.title('Vote Average Distribution')
plt.xlabel('Vote Average')
plt.ylabel('Frequency')
plt.show()
```

```python
#Conclusion: The histogram displays the distribution of vote averages across all movies. This helps in understanding how movie ratings are spread out. A normal distribution might indicate a balanced rating system, whereas a skewed distribution could suggest a bias in the ratings.
```

```python
#d. Count of Movies by Runtime

plt.figure(figsize=(10, 6))
sns.countplot(x=df['runtime'])
plt.title('Count of Movies by Runtime')
plt.xlabel('Runtime (minutes)')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
#Conclusion: This count plot shows how many movies fall into various runtime categories. It helps in identifying common lengths of movies and whether most movies are short, medium, or long in duration.
```

```python
#Boxplot of Vote Average by Original Language

plt.figure(figsize=(8, 6))
sns.boxplot(x='original_language', y='vote_average', data=df)
plt.title('Boxplot of Vote Average by Original Language')
plt.xlabel('Original Language')
plt.ylabel('Vote Average')
plt.show()
```

```python
#Conclusion: The boxplot visualizes the distribution of vote averages across different languages. It highlights the median rating, the spread of ratings, and potential outliers. This can be used to analyze whether certain languages tend to receive higher or lower ratings.
```

```python
#Heatmap of Vote Average by Genres and Original Language

pivot_table = df.pivot_table(index="genres", columns="original_language", values="vote_average", aggfunc='mean')

plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, annot=True, cmap='YlGnBu', fmt='.1f')
plt.title('Heatmap of Vote Average by Genres and Original Language')
plt.xlabel('Original Language')
plt.ylabel('Genres')
plt.show()
```

```python
#Conclusion: The heatmap provides a visual representation of the average ratings for each genre-language combination. It helps in identifying trends or correlations between genres and languages, such as whether movies in a particular genre and language tend to receive higher ratings.
```
